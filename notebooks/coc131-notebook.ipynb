```python
# Test q2 with a sample from your dataset
import numpy as np
import matplotlib.pyplot as plt

# Get a subset of the data for testing
sample_data = coc.x[:100]  # Take 100 samples

# Standardize the data
standardized_data, scaler = coc.q2(sample_data)

# Create a visualization to show the effect of standardization
plt.figure(figsize=(15, 6))

# Plot original data distribution
plt.subplot(1, 2, 1)
plt.hist(sample_data.flatten(), bins=50, alpha=0.7)
plt.title('Original Data Distribution')
plt.xlabel('Pixel Values')
plt.ylabel('Frequency')
plt.grid(alpha=0.3)
plt.axvline(np.mean(sample_data), color='r', linestyle='dashed',
            label=f'Mean: {np.mean(sample_data):.2f}')
plt.axvline(np.mean(sample_data) + np.std(sample_data), color='g', linestyle='dashed',
            label=f'Mean + Std: {np.mean(sample_data) + np.std(sample_data):.2f}')
plt.axvline(np.mean(sample_data) - np.std(sample_data), color='g', linestyle='dashed')
plt.legend()

# Plot standardized data distribution
plt.subplot(1, 2, 2)
plt.hist(standardized_data.flatten(), bins=50, alpha=0.7)
plt.title('Standardized Data Distribution (std=2.5)')
plt.xlabel('Standardized Values')
plt.ylabel('Frequency')
plt.grid(alpha=0.3)
plt.axvline(np.mean(standardized_data), color='r', linestyle='dashed',
            label=f'Mean: {np.mean(standardized_data):.2f}')
plt.axvline(np.mean(standardized_data) + 2.5, color='g', linestyle='dashed',
            label=f'Mean + 2.5: {np.mean(standardized_data) + 2.5:.2f}')
plt.axvline(np.mean(standardized_data) - 2.5, color='g', linestyle='dashed')
plt.legend()

plt.tight_layout()
plt.savefig("../visualizations/q2-1.png", dpi=300, bbox_inches='tight')
plt.show()

# Verify the standard deviation is 2.5
print(f"Standard deviation of original data: {np.std(sample_data):.4f}")
print(f"Standard deviation of standardized data: {np.std(standardized_data):.4f}")
print(f"Mean of original data: {np.mean(sample_data):.4f}")
print(f"Mean of standardized data: {np.mean(standardized_data):.4f}")
```
